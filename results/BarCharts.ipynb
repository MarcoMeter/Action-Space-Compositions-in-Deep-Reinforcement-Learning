```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.patches as mpatches
```

```python
# SB

x = ['Multiagent Small', 'Multiagent Full', 'Bucket 201', 'Bucket 41', 'Bucket 21', 'Bucket 11', 'Bucket 7', 'Threshold']
y = [110, 113, 120, 67, 62, 62, 57, 57]

plt.clf()
fig, ax = plt.subplots()
ax.set_xlim(0, 135)
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
ax.barh(ind, y, width, color="lightblue")
ax.set_yticks(ind)
ax.set_yticklabels(x, minor=False)
plt.xlabel('Minutes')
for i, v in enumerate(y):
    ax.text(v + 3, i - 0.1, str(v), color='black', fontweight='regular')
plt.show()
fig.savefig('./plots/sb_time.pdf', bbox_inches='tight')
```

```python
# BRO Thresolds

blue_patch = mpatches.Patch(color='lightblue', label='Inner')
red_patch = mpatches.Patch(color='lightcoral', label='Outer')


x = ['0.9', '0.8', '0.67', '0.5', '0.33', '0.2', '0.1', '0.9', '0.8', '0.67', '0.5', '0.33', '0.2', '0.1']
y = [337, 345, 355, 359, 365, 374, 392, 411, 411, 407, 401, 381, 373, 353]
colors = ['lightcoral'] * 7 + ['lightblue'] * 7


plt.clf()
fig, ax = plt.subplots()
ax.set_xlim(0, 475)
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
ax.barh(ind, y, width, color=colors)
ax.set_yticks(ind)
ax.set_yticklabels(x, minor=False)
ax.barh(ind, y, width, color=colors)
ax.set_yticks(ind)
ax.set_yticklabels(x, minor=False)
plt.xlabel('Minutes')
for i, v in enumerate(y):
    ax.text(v + 3, i - 0.1, str(v), color='black', fontweight='regular')
ax.legend(handles=[blue_patch, red_patch], loc='lower right')
plt.show()
fig.savefig('./plots/bro_threshold_time.pdf', bbox_inches='tight')
```
