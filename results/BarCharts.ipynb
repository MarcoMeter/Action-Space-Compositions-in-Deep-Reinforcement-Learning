```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.patches as mpatches
```

```python
x = ['Multiagent', 'Bucket 41', 'Bucket 11', 'Bucket 3', 'Threshold 0.5', 'Threshold 0.1', 'Threshold 0.0']
sb_y = [10, 6.9, 6, 5.8, 5.8, 5.9, 5.8]
bro_y = [26.7, 23.3, 22.7, 22, 22, 21, 21.7]
 
# create plot
fig, ax = plt.subplots()
ax.set_xlim(0, 29)
width = 0.75 # the width of the bars
ind = np.arange(len(bro_y))  # the x locations for the groups
ax.barh(ind, bro_y, width, color="lightblue")
ax.barh(ind, sb_y, width, color="lightcoral")
ax.set_yticks(ind)
ax.set_yticklabels(x, minor=False)
plt.xlabel('Hours')
for i, v in enumerate(bro_y):
    ax.text(v + 0.125, i - 0.1, str(v), color='black', fontweight='regular')
for i, v in enumerate(sb_y):
    ax.text(v + 0.125, i - 0.1, str(v), color='black', fontweight='regular')
red_patch = mpatches.Patch(color='lightcoral', label='SB')
blue_patch = mpatches.Patch(color='lightblue', label='BRO')
plt.legend(handles=[red_patch, blue_patch], loc='center right')
plt.show()
fig.savefig('./plots/time.pdf', bbox_inches='tight')
```
